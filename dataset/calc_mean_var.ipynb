```python
import os
import numpy as np
from PIL import Image
```

```python
root = "dataset"
file_ending = 'png'
length_root = len(os.listdir(root))
```

```python
mean_red_list = []
mean_green_list = []
mean_blue_list = []

var_red_list = []
var_green_list = []
var_blue_list = []

total_pixels = 0

# Iterate through all directories
for i in range(len(os.listdir(root))):
    directory = i + 1
    print("Processing directory", directory)
    image_dir = os.path.join(root, str(directory))

    # Get image files
    image_files = [fn for fn in os.listdir(image_dir) if file_ending in fn]

    for image_name in image_files:
        # Load the image
        img = Image.open(os.path.join(image_dir, image_name))
        
        # Convert the image to numpy array (height x width x 3)
        img_np = np.array(img) / 255.0  
        
        # Get the number of pixels in the image
        num_pixels = img_np.shape[0] * img_np.shape[1]
        total_pixels += num_pixels

        # Split the image into R, G, and B channels
        red = img_np[:, :, 0].ravel()
        green = img_np[:, :, 1].ravel()
        blue = img_np[:, :, 2].ravel()

        # Calculate the mean and variance for each channel for this image
        mean_red = np.mean(red)
        mean_green = np.mean(green)
        mean_blue = np.mean(blue)

        var_red = np.var(red)
        var_green = np.var(green)
        var_blue = np.var(blue)

        # Append the mean and variance values for each image
        mean_red_list.append(mean_red)
        mean_green_list.append(mean_green)
        mean_blue_list.append(mean_blue)

        var_red_list.append(var_red)
        var_green_list.append(var_green)
        var_blue_list.append(var_blue)

# Calculate mean of means for each channel
mean_red_overall = np.mean(mean_red_list)
mean_green_overall = np.mean(mean_green_list)
mean_blue_overall = np.mean(mean_blue_list)

# Calculate variance of variances for each channel
var_red_overall = np.mean([v + (m - mean_red_overall)**2 for v, m in zip(var_red_list, mean_red_list)])
var_green_overall = np.mean([v + (m - mean_green_overall)**2 for v, m in zip(var_green_list, mean_green_list)])
var_blue_overall = np.mean([v + (m - mean_blue_overall)**2 for v, m in zip(var_blue_list, mean_blue_list)])

# Print the overall mean and variance for each channel
print(f"Mean values - R: {mean_red_overall}, G: {mean_green_overall}, B: {mean_blue_overall}")
print(f"Variance values - R: {var_red_overall}, G: {var_green_overall}, B: {var_blue_overall}")
```
