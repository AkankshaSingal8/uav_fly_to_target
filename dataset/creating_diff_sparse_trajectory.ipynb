```python
import os
from typing import Iterable, Dict

import csv

import numpy as np
from matplotlib.image import imread
from tqdm import tqdm
from PIL import Image

import seaborn as sns
import pandas as pd
import scipy.io

import cv2
```

```python

file_ending = 'png'
file_name = "data_out.csv"

original_coreset = 'coreset'
result_directory = "diff_coreset"
```

```python
# Creating sparse subset of the trajectory
length_dest = len(os.listdir(result_directory))

for i in range(len(os.listdir(original_coreset))):
    directory = i + 1
    print("Processing directory", directory)
    
    # print("Print length dest", length_dest)
    
    new_directory_path = f"{result_directory}/{directory}"
    if not os.path.exists(new_directory_path):
        print(new_directory_path)
        os.makedirs(new_directory_path)
            
    
    csv_file_name = f"{original_coreset}/{str(directory)}/{file_name}"
    labels = np.genfromtxt(csv_file_name, delimiter=',', skip_header=1, dtype=np.float32)

    csv_file_name = f"{original_coreset}/{str(directory)}/{file_name}"
    labels = np.genfromtxt(csv_file_name, delimiter=',', skip_header=1, dtype=np.float32)
    
    image_files = [fn for fn in os.listdir(f"./{original_coreset}/{str(directory)}") if file_ending in fn]
    n_images = len(image_files)

    # Load the goal images
    goal_image = cv2.imread(f'{original_coreset}/{str(directory)}/Image{n_images}.png')

    for i in range(1, n_images + 1):
        # Load the current image
        current_image = cv2.imread(f'{original_coreset}/{str(directory)}/Image{i}.png')

        # Compute the absolute difference between the images
        difference = cv2.absdiff(current_image, goal_image)

        # Save the result
        print(f'{new_directory_path}/Image{i}.png')
        cv2.imwrite(f'{new_directory_path}/Image{i}.png', difference)
    
    columns = ['vx', 'vy', 'vz', 'omega_z']
    df = pd.DataFrame(labels, columns=columns)
    print(f"{new_directory_path}/{file_name}")
    df.to_csv(f"{new_directory_path}/{file_name}", index=False)
```
